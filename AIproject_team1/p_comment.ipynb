```
# 先處理標題和內容 小寫, 空格去除, 再去除跳脫字元
df = pd.read_csv("./drive/My Drive/機不可失/nier/ptt_label/ptt_samsung_ohe.csv")
for k in range(len(df["標題"])):
    df.iloc[k, 0] = (df.iloc[k, 0]).lower()
    df.iloc[k, 0] = (df.iloc[k, 0]).replace(" ", "")

for k in range(len(df["內容"])):
    df.iloc[k, 1] = (df.iloc[k, 1]).lower()
    df.iloc[k, 1] = (df.iloc[k, 1]).replace(" ", "")
def jana(s):
  s = s.lower()
  s = s.replace("\r", "").replace("\n", "").replace(" ", "")
  s = s.strip()
  return s
df["內容"] = df["內容"].apply(jana)
df
```
