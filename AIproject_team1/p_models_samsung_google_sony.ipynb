```python
# 型號(標題)
for i in range(len(df)):
    model_list = [] # 每跑一次做一個新的 list
    for m in model_samsung.keys(): # 用 m 去跑 model_samsung的 key
        # model_dict[m]是拿出value值, 如果有任何的字典內的value值在標題內
        if any(t in df["標題"][i] for t in model_samsung[m]): 
            # if m in model_list:
            #     break
            # else:
          model_list.append(m)
    models = ", ".join(model_list) # 把 model_list 變成字串並用,分隔
    df.iloc[i, 0] = models #指定第i列的第0欄位子放 models 這個的字串

# 型號(內容)
for i in range(len(df)):
    model_list = []
    for m in model_samsung.keys():
        if any(t in df["內容"][i] for t in model_samsung[m]):
            if m in model_list:
                break
            else:
                model_list.append(m)
    models = ", ".join(model_list)
    df.iloc[i, 1] = models
     
df
```
