```python
import mtcnn
from mtcnn import MTCNN
import cv2
import matplotlib.pyplot as plt
import numpy as np
from glob import glob
```

```python
path = '/content/drive/My Drive/week10/face_detection/5celeb/data/*/*/*.jpg'
```

```python
img_paths = glob(path)
len(glob(path))
```

```python
# create the detector, using default weights
detector = MTCNN()
```

```python
img = cv2.imread(img_paths[0])[:,:,::-1] # read image
# detect face
results = detector.detect_faces(img)
print(results)
# extract the bounding box from the first face
x1, y1, width, height = results[0]['box']
x2, y2 = x1 + width, y1 + height
print((x1, y1), (x2, y2), img.shape)
plt.imshow(img)
```

```python
for path in img_paths[:5]:
    img = cv2.imread(path)[:,:,::-1] # read image
    # detect face
    results = detector.detect_faces(img)
    # extract the bounding box from the first face
    assert len(results) >= 1
    x1, y1, width, height = results[0]['box']
    x2, y2 = x1 + width, y1 + height
    print(x1, x2, y1,y2)
    # draw bbox
    img = cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 5).get()
    # crop face
    patch = img[max(y1, 0):y2, max(x1, 0):x2]
    # show image
    plt.imshow(patch)
    plt.show()
    plt.imshow(img)
    plt.show()
```
