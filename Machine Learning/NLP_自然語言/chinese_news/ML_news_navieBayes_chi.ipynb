#### 處理資料夾內有大量子資料夾情況時使用 glob, os
#### glob 類似 windows 下的文件搜索, 搜索完會返回一個 list 

```
import glob, os
import pandas as pd
fname = glob.glob("./drive/My Drive/chinese_news_trans/*/*.txt")
tname = glob.glob("./drive/My Drive/chinese_news_test/*/*.txt")
fname
```

#### os.path.split() 方法, 把檔案路徑切開, 會從最後面的 / 開始切
#### 切完之後會是一個 truple, 拿取檔案名稱當作 target的時候好用
#### 參考：https://blog.csdn.net/zzc15806/article/details/81352742

```
import  os
dn = os.path.split("./drive/My Drive/chinese_news_trans/教育/3127.txt")
print(dn)
print(dn[0])
dn2 = dn[0]
d = os.path.split(dn2)
print(d)
print(d[-1])
```

```
# 訓練資料整理成 DataFrame
datas = {
    "content":[],
    "target":[]
}
for i in fname:
  with open(i, mode="r", encoding="utf-8") as f:
    artical = f.read()
    datas["content"].append(artical)
    dn = os.path.split(i)[0]
    dn = os.path.split(dn)[-1]
    datas["target"].append(dn)
train_df = pd.DataFrame(datas)
train_df
```

```
# 測試資料處理成 DataFrame
datas = {
    "content":[],
    "target":[]
}
for i in tname:
  with open(i, mode="r", encoding="utf-8") as f:
    artical = f.read()
    datas["content"].append(artical)
    dn = os.path.split(i)[0]
    dn = os.path.split(dn)[-1]
    datas["target"].append(dn)
test_df = pd.DataFrame(datas)
test_df
```

```
# 將 target 處理成數字
u = train_df["target"].unique()
trans = {name:i for i, name in enumerate(u)}
r_trans = {i:name for i, name in enumerate(u)}
y_train = train_df["target"].replace(trans)
y_test = test_df["target"].replace(trans)
print(u)
print(trans)
print(r_trans)
```

#### jieba 繁體大辭典:  https://github.com/fxsjy/jieba/raw/master/extra_dict/dict.txt.big
#### jieba 參考: https://github.com/fxsjy/jieba

```
import jieba
from urllib.request import urlretrieve
if not os.path.exists("dict.txt.big"):
  print("dict.txt.big 不存在, 下載")
  url = "https://github.com/fxsjy/jieba/raw/master/extra_dict/dict.txt.big"
  urlretrieve(url, "dict.txt.big")
else:
  print("dict.txt.big 已經存在")

# 載入大辭典後切割文字
jieba.set_dictionary("dict.txt.big")
def newscut(s):
  s = " ".join(jieba.cut(s))
  s = s.replace("\n", "").replace("\r", "")
  return s
x_train = train_df["content"].apply(newscut)
x_test = test_df["content"].apply(newscut)
x_train
```

```
# 資料處理成標籤及出現次數
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
x_train_vec = vec.fit_transform(x_train)
x_test_vec = vec.transform(x_test)
```

```
# 查看字詞出現次數
vec.vocabulary_
```

```
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train_vec, y_train)
```

```
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test_vec)
print("預測率為:", accuracy_score(y_test, pre))
```

```
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, pre)
p = ["預測 {}".format(i) for i in u]
r = ["真實 {}".format(i) for i in u]
pd.DataFrame(mat, columns=p, index=r)
```

```
# 應用於預測
p = input("輸入一篇新聞:")
p = vec.transform([newscut(p)])
ans = clf.predict(p)[0]
print("分類應該是:", r_trans[ans])
print() # 分隔
proba = clf.predict_proba(p)[0]
for n, prob in sorted(zip(u, proba), key=lambda x:x[1], reverse=True):
  print(n, "的機率是", prob * 100)
```

```
# zip的用法 將兩個 list的東西一樣的位子連結起來
A = [1, 2, 3]
B = [4, 5, 6]
C = [7, 8, 9, 10]
print(list(zip(A, B, C)))
```

```
# sorted 是依照第一個位子來排序
# 設定一個排序函式, 用第一個位子來排序
def s(x):
  retrun x[1]
sorted(zip(u, proba), key=s)

# 使用 lamdba 處理上述, lambda x:x[1] 傳入x, 回傳 x[1]
sorted(zip(u, proba), key=lambda x:x[1], reverse=True)
```

#### News 主要處理的幾個項目
 1. glob.glob: 類似 windows 的搜尋功能, 可以把資料夾下面的東西全部秀出
 2. os.path.split: 切割檔案的路徑, 用最後一個 \ 切開, 用於檔案資料夾當作是 target 時
 3. replace: 用於將 target 的文字換成數字 y_train = train_df["target"].replace(trans)
 4. predict_proba: 預測後會將機率從高到低用list表示
 5. zip: 將兩個 list 相同 index的東西放一起
 6. lambda: 無名函式, x->相當於 def s(x): 的 x, x[1]->相當於 retrun x[1]
