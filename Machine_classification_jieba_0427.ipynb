```python
#正式開始
import pandas as pd
train_df = pd.read_csv("poem_train.csv", encoding="utf-8")
train_df
```

```python
test_df = pd.read_csv("poem_test.csv", encoding="utf-8")
test_df
```

```python
#先處理答案:許含式庫處理成數字
#pandas 的 replace功能, 可以直接做替換
#tarns = {"李白":0, "杜甫":1, "白居易":2}
u = train_df["作者"].unique()
trans = {name: i for i, name in enumerate(u)}
reverse_trans = {i:name for name, i in trans.items()}
print(trans)
print(reverse_trans)
y_train = train_df["作者"].replace(trans)
y_test = test_df["作者"].replace(trans)
y_test
```

```python
#切割字詞, \r \n處理掉, 切字,如果這是詩詞分類, 標點符號應該去掉, 因為他不影響結果 
import jieba
s = "平林漠漠煙如織，寒山一帶傷心碧。\r\n暝色入高樓，有人樓上愁。玉階空佇立，宿鳥歸飛急。"
s = " ".join(jieba.cut(s))
s.replace("\r", "").replace("\n", "")
```

```python
def poemcut(s):
    s = " ".join(jieba.cut(s))
    s = s.replace("\r", "").replace("\n", "")
    return s
# pandas表格: 多行多列 DataFrame, 單行單列 Series, 這邊是用 Series的 apply(功能名字) ->打名字就好, 之後會自己帶 
x_train = train_df["內容"].apply(poemcut)
x_test = test_df["內容"].apply(poemcut)
x_test
```

```python
# 訓練資料: 用 fit, trans  ; 測試資料: 完全根據訓練資料暨有的欄位去 trans
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
#測試資料不用fit, 訓練資料才要fit, transform
x_train_vec = vec.fit_transform(x_train)
x_test_vec = vec.transform(x_test)
```

```python
#print(x_train_vec)
vec.vocabulary_
#檢查有沒有把標點符號去掉用下面的方法, "平林" 在14684個欄位, 把，或是。待入下面[]裏面, 如果找不到就代表有去掉
vec.vocabulary_["平林"]
```

```python
from sklearn.naive_bayes import MultinomialNB
#單純貝式運算
clf = MultinomialNB()
clf.fit(x_train_vec, y_train)
```

```python
#算正確率
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test_vec)
print("正確率:", accuracy_score(y_test, pre))
```

```python
#混淆矩陣
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, pre)
c = ["{}(預測)".format(name) for name in u]
i = ["{}(原本)".format(name) for name in u]
pd.DataFrame(mat, columns=c, index=i)
```

```python
#回顧
p =input("輸入一首詩:")
#注意[], [第一首詩, 第二首詩]
p = vec.transform([poemcut(p)])
#注意[], [第一個答案, 第二個答案]
ans = clf.predict(p)[0]
print("答案應該是:", reverse_trans[ans])
```
