```python
# glob 用法
import glob, os
import pandas as pd
fnames = glob.glob("chinese_news_trans/*/*.txt")
datas= {
    "content": [],
    "target": []
}
for fn in fnames:
    with open(fn, "r", encoding="utf-8") as f:
        content = f.read()
        datas["content"].append(content)
        dn = os.path.split(fn)[0]
        dn = os.path.split(dn)[-1]
        datas["target"].append(dn)
train_df = pd.DataFrame(datas)
train_df
#正確率, 混亂矩陣, 載入新的新聞
```

```python
fnames = glob.glob("chinese_news_test/*/*.txt")
datas= {
    "content": [],
    "target": []
}
for fn in fnames:
    with open(fn, "r", encoding="utf-8") as f:
        content = f.read()
        datas["content"].append(content)
        dn = os.path.split(fn)[0]
        dn = os.path.split(dn)[-1]
        datas["target"].append(dn)
test_df = pd.DataFrame(datas)
test_df
```

```python
import jieba
#載入大辭典
jieba.set_dictionary("dict.txt.big")
def newscut(s):
    s = " ".join(jieba.cut(s))
    s = s.replace("\r", "").replace("\n", "")
    return s
# pandas表格: 多行多列 DataFrame, 單行單列 Series, 這邊是用 Series的 apply(功能名字) ->打名字就好, 之後會自己帶 
x_train = train_df["content"].apply(newscut)
x_test = test_df["content"].apply(newscut)
print(x_train)
print(x_test)
```

```python
#將target轉換成數字
u = train_df["target"].unique()
trans = {name: i for i, name in enumerate(u)}
reverse_trans = {i:name for name, i in trans.items()}
print(trans)
print(reverse_trans)
y_train = train_df["target"].replace(trans)
y_test = test_df["target"].replace(trans)
y_test
```

```python
#把每一個詞轉換成數量, 每一個詞出現幾遍
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
x_train_vec = vec.fit_transform(x_train)
x_test_vec = vec.transform(x_test)
```

```python
#單純貝式運算
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train_vec, y_train)
```

```python
#預測, 算機率
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test_vec)
print("正確率:", accuracy_score(y_test, pre))
```

```python
#混淆矩陣
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, pre)
c = ["{}(預測)".format(name) for name in u]
i = ["{}(原本)".format(name) for name in u]
pd.DataFrame(mat, columns=c, index=i)
```
