```python
# kaggle, house price :https://www.kaggle.com/c/house-prices-advanced-regression-techniques/overview
import pandas as pd
train_df = pd.read_csv("h_train.csv", encoding="utf-8")
train_df
```

```python
test_df = pd.read_csv("h_test.csv", encoding="utf=8")
test_df
```

```python
# 將 train_df 和 test_df 合併成為一個 DataFrame, 叫做 datas
train_df_drop = train_df.drop(["SalePrice"], axis=1)
datas = pd.concat([train_df_drop, test_df], axis=0)
datas = datas.drop(["Id"], axis=1)
datas
```

```python
#找出缺失值, 並統計
na = datas.isna().sum()
# 拿出 na(也就是缺失值)裡面不是0 也就是 True的人留下, 是 False的人去掉,並由大排到小
na[na != 0].sort_values(ascending=False)
datas = datas.drop(na[na > 1000].index, axis=1)
datas
```

```python
# 填補缺失值(數字部分), MSSubClass是屬於類別 要先去掉別填
med = datas.median().drop(["MSSubClass"])
datas = datas.fillna(med)
```

```python
# one-hot encoding,
datas = pd.get_dummies(datas)
datas = pd.get_dummies(datas, columns=["MSSubClass"])
datas
```

```python
train_df.shape
#是一個 tuple, 拿取第一個index,
train_df.shape[0]
```

```python
#資料處理完成, 分開資料
x_train = datas.iloc[:train_df.shape[0]] #訓練資料
y_train = train_df["SalePrice"] #訓練資料答案
x_predict = datas.iloc[train_df.shape[0]:] #測試資料
predict_id = test_df["Id"] #ID
```

```python
# RandomForestregressr, 用 GridSearchCV 找出最佳參數
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
reg = RandomForestRegressor()
params = ({
    "n_estimators":range(20, 100),
    "max_depth":range(6, 20)
})
grid = GridSearchCV(reg, params, verbose=1, cv=10, n_jobs=4)
grid.fit(x_train, y_train)
print("最佳參數:", grid.best_params_)
print("最佳分數:", grid.best_score_)
```

```python
reg = RandomForestRegressor(n_estimators=72, max_depth=16)
reg.fit(x_train, y_train)
pre = reg.predict(x_predict)
result = pd.DataFrame({
    "ID":predict_id,
    "SalePrice":pre
})
result.to_csv("house_price_randomforest.csv", encoding="utf-8", index=False)
result
```
