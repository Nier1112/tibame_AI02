```python
count = 1
pos = 0
neg = 0
for line in open("drive/My Drive/phone_c3.txt",encoding='utf-8', newline=""):
    print(f"處理第{count}筆")
    # if count == 30001:
    #   break
    # line = line.lower()
    line = line.strip('\n')
#     print(line)
    try:
        test_text = [line]
        input_dict = {"text": test_text}
        results = senta.sentiment_classify(data=input_dict)
    except:
        continue
#     print(s.sentiments)
    if results[0]['positive_probs'] > 0.9:
        with open("drive/My Drive/pos.txt", "a", encoding="utf-8",newline="") as f:
            f.write(line + "\n")
        pos += 1
    elif results[0]['positive_probs'] < 0.1:
        with open("drive/My Drive/neg.txt", "a", encoding="utf-8",newline="") as f:
            f.write(line + "\n")        
        neg += 1
    count += 1 
print(pos, neg)
```
